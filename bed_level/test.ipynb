```python
#Test
import matplotlib.pyplot as plt
import numpy as np

```

```python
"""
350 x 350mm and a Z-axis stretching to 400mm
"""
length_width =350.
spacing = length_width /5.
offset = spacing/2
xlist = np.linspace(offset,length_width - offset, 4)
ylist = np.linspace(offset,length_width - offset, 4)
```

```python
X, Y = np.meshgrid(xlist, ylist)

Z = [[-0.924,	0.387	,1.062	,2.762],
    [-1.562,	0	    ,0.887	,2.575],
    [-2.749,	-0.924	,0.487	,2.95],
    [-4.274,	-2.149	,-0.187	,3.7]]
```

```python
fig,ax=plt.subplots(1,1,figsize = [4,4])

N=30
cp = ax.contourf(X, Y, Z, N, cmap="viridis")
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Bed_profile_24062023', fontsize = 20)
ax.set_ylabel('y (mm)', fontsize = 20)
ax.set_xlabel('x (mm)', fontsize = 20)
plt.show()
```
